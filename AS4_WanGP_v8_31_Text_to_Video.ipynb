```python
#for A100 80 GB VRAM, 160 GB System ram to run multiple batches
import subprocess
from concurrent.futures import ThreadPoolExecutor

# మీ దగ్గర ఉన్న 17 బ్యాచ్ ఫైల్స్ పేర్లు ఇక్కడ ఇవ్వండి
batch_files = [f"Batch_{i}.zip" for i in range(14, 26)]

def run_batch(batch_name):
    print(f"Starting: {batch_name}")
    # కమాండ్‌ను రన్ చేస్తుంది
    subprocess.run(["python", "wgp.py", "--process", batch_name])
    print(f"Finished: {batch_name}")

# max_workers=5 అంటే ఒకేసారి 5 బ్యాచ్‌లు మాత్రమే రన్ అవుతాయి.
# ఒకటి అవ్వగానే 6వ బ్యాచ్ ఆటోమేటిక్‌గా స్టార్ట్ అవుతుంది.
with ThreadPoolExecutor(max_workers=5) as executor:
    executor.map(run_batch, batch_files)
```
